# 1. Read the dataset to python environment

```python
import pandas as pd
iris_data=pd.read_excel('iris.xls')
iris_data
```

The Iris Dataset contains four features (length and width of sepals and petals(SL,SW,PL,PW) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor).. These measures were used to classify the species.


# 2. Display the columns in the dataset.

```python
print(iris_data.columns)
```

# 3. Calculate the mean of each column of the dataset.

```python
SL_mean=iris_data['SL'].mean()
print('Mean of SL is',SL_mean)
SW_mean=iris_data['SW'].mean()
print('Mean of SW is',SW_mean)
PL_mean=iris_data['PL'].mean()
print('Mean of PL is',PL_mean)
PW_mean=iris_data['PW'].mean()
print('Mean of PW is',PW_mean)

```

Mean and Median of Sepal length,Sepal Width,Petal Length and Petal Width based on the species of iris is illustrated above.

```python
iris_data.groupby('Classification').agg(['mean', 'median'])
```

# 5. visualizations using the dataset. 

# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


sns.lmplot(x="SL", y="SW", data=iris_data, hue="Classification", fit_reg=False, legend=False)
plt.xlabel('sepal_length(cm)')
plt.ylabel('sepal_width(cm)')
plt.title('Scatter Plot of Sepal Length and Width')
plt.legend()
plt.show()


```

From the above plot we can observe that setosa is the species with longer sepal width and length .Whereas versicolor and virginica have sepals with almost same size

```python
sns.lmplot(x="PL", y="PW", data=iris_data, hue="Classification", fit_reg=False, legend=False)
plt.xlabel('Petal_length(cm)')
plt.ylabel('Petal_width(cm)')
plt.title('Scatter Plot of Petal Length and Width')
plt.legend()
plt.show()
```

From the above figure it is clear that setosa has the smallest petals .
The range of it's petal length varies from 0.5 to 2 cm and width varies from 0.2 to 0.7 cm
Versicolor is larger than setosa but it's petal length is smaller than that of virginica.

# Heat Map

```python
sns.heatmap(iris_data.corr(),annot=True)
```

Numerical feautures in the above illustration gives an idea about how variables are corelated.
Main findings are:
    1. Sepal Length(SL) is highly corelated with Petal Length and Width (PL and PW) .
    2. A slight inverse corelation can be observed between Sepal Width and Lenth(SW and SL) .Sepal Width is also inversly 
    corelated with Petal Width and Length(PW and PL) .
    3.Corelation between Petal Width and Petal length is remarkable with a value of 0.96.
   

# Box Plot

```python
sns.set(style="ticks") 
plt.figure(figsize=(8,10))
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=iris_data)
plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=iris_data)
plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=iris_data)
plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PL',data=iris_data)
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.violinplot(x='Classification',y='SL',data=iris_data)
plt.subplot(2,2,2)
sns.violinplot(x='Classification',y='SW',data=iris_data)
plt.subplot(2,2,3)
sns.violinplot(x='Classification',y='PL',data=iris_data)
plt.subplot(2,2,4)
sns.violinplot(x='Classification',y='PW',data=iris_data)
plt.show()
```
